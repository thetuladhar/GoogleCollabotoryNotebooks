In 2020, I rekindled my interest in collecting Pokemon cards, and only recently discovered that some of the cards in my collection are rare and have significant market value. While I had previously checked the prices of a few cards, the process quickly became tedious, as I had to look up each card individually. To simplify the task and combine my hobby with my web scraping skills, I created a database that provides real-time values for my entire collection without the need to navigate multiple websites.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from IPython.core.display import HTML
```

```python
#the base url of the website that Iam using
base='https://www.pokemonwizard.com/'

#function to scrape necessary information from URL
def add(url):
  #request the page 
  page= requests.get(url)
  soup=BeautifulSoup(page.content,"html.parser")
  #scrape the necessary info
  header=soup.find_all("h1",class_="h1-header")[0].get_text()
  image=soup.find("img",class_="img-fluid lozad")["src"]
  table=soup.find_all("table",class_="table table-borderless")
  current=table[0].find_all("span")[0].get_text().split(" ")[0]
  updown=table[0].find_all("span")[0].get_text().split(" ")[1]
  last7=table[0].find_all("span")[2].get_text().replace("\n","")
  last30=table[0].find_all("span")[3].get_text().replace("\n","")
  YTD=table[0].find_all("span")[-1].get_text().replace("\n","")

  #fix the image bug
  #some images were read in from defferent website
  if image[:5]=="https":
    picture=image
  else:
    picture=base+image
  
  #Initiating the data
  data = [{"Name": header,
          "Picture": picture,
          'Market_Price': current,
          'up/down': updown,
          'last7': last7,
          'last30': last30,
          'YTD': YTD}]
  return data

#function to convert text to html path
def path_to_image_html(path):
    return '<img src="'+ path + '" width= "60" >'
```

```python
#ideally this would be read from a CSV file but for purposes of DEMO it a list
collection=['https://www.pokemonwizard.com/cards/477057/mewtwo-vstar',
    'https://www.pokemonwizard.com/cards/478103/pikachu-secret',
     'https://www.pokemonwizard.com/cards/124026/charizard',
     'https://www.pokemonwizard.com/cards/250295/mew-secret',
     'https://www.pokemonwizard.com/cards/251102/pikachu-v-swsh145',
     'https://www.pokemonwizard.com/cards/251103/poke-ball-swsh146',
     ]
#collection=list(pd.read_csv('/content/collection.csv')['links'])
```

```python
df=pd.DataFrame()
for i in collection:
  data=add(i)
  df=df.append(data,ignore_index=True)
#df.sort_values(by="Market_price")
#currently the values are in string format
#for future version(s) I would like to convert the values into floats 
```

```python
HTML(df.to_html(index=False,escape=False,formatters=dict(Picture=path_to_image_html)))
```

```python
prices=df["Market_Price"]

sum=0
for i in prices:
  sum=sum+float(i[1:])
print("Your Total Value of the collection as of Today is $",round(sum,2))
```
