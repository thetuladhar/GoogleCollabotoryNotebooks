```python
import pandas as pd
```

```python
url="/content/drive/MyDrive/World Energy Consumption.csv"
df =pd.read_csv(url)
#df.columns.to_list()
```

```python
#df["country"].unique().tolist()
```

```python
country=df["country"]=='United States'
years=df["year"]>=1995
country_data = df[country & years].sort_values(by="year", ascending=False)
#country_data['population']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Columns to plot
columns_to_plot = ["coal_production", "oil_production", "gas_production"]

# Set the style for the plot (optional)
sns.set(style="darkgrid")

# Create a line plot for each selected column
for column in columns_to_plot:
    sns.lineplot(data=country_data, x="year", y=column, label=column)

# Add labels and a title
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Production vs. Year in United States")

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plot (optional)
sns.set(style="darkgrid")

# Create a line plot for GDP over the years
sns.lineplot(data=country_data, x="year", y="gdp", label="GDP")

# Add labels and a title
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GDP Over the Years in United States")
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


filtered_data = df[df['country'].isin(["United States", "China"])]


pivot_data = filtered_data.pivot(index='year', columns='country', values='gdp')


plt.figure(figsize=(10, 6))
plt.plot(pivot_data.index, pivot_data["United States"], label="United States")
plt.plot(pivot_data.index, pivot_data["China"], label="China")


plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("US vs. China GDP Over Time")
plt.legend()

# Show the line plot
plt.show()
```

```python
#selected_countries = ["United States", "China", "India", "Japan", "Nepal"]
selected_countries = ["Nepal"]
filtered_data = df[df['country'].isin(selected_countries)& (df['year'] >= 1900)]

# Pivot the data to create a line plot
pivot_data = filtered_data.pivot(index='year', columns='country', values='population')

# Create a line plot
plt.figure(figsize=(10, 6))
for country in selected_countries:
    plt.plot(pivot_data.index, pivot_data[country], label=country)

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Comparison for Selected Countries")
plt.legend()


from matplotlib.ticker import FuncFormatter
# Define a function to format y-axis labels with thousands separators
def thousands_format(x, pos):
    return '{:,.0f}'.format(x)

# Apply the custom y-axis label formatting
formatter = FuncFormatter(thousands_format)
plt.gca().yaxis.set_major_formatter(formatter)

# Show the line plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px

#filter of selected countries
selected_countries = ["United States", "China", "India", "Japan", "Nepal"]
filtered_data = df[(df['country'].isin(selected_countries)) & (df['year'] >= 1900)]

#plot using plotly
fig = px.line(filtered_data, x='year', y='population', color='country',
              labels={'year': 'Year', 'population': 'Population', 'country': 'Country'},
              title="Population Comparison for Selected Countries (1995 Onwards)")

# Hover Function
fig.update_traces(mode='lines+markers', hovertemplate='%{y:,.0f} Population')

#show
fig.show()
```
